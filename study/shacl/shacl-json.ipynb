# SHACL JSON experiment

```python
from pyshacl import validate
import rdflib
from rdflib import SH, XSD
```

## Shapes

```python
shapesJson = """
{
"@context": {
    "sh": "http://www.w3.org/ns/shacl#",
    "xsd": "http://www.w3.org/2001/XMLSchema#"
  },
    "@graph": [
        {
      "@id": "https://good.vibrations.dev/PromptTemplateShape",
      "@type": "sh:NodeShape",
      "sh:targetClass": {
        "@id": "https://good.vibrations.dev/PromptTemplate"
      },
      "sh:property": [
        {
          "sh:datatype": {
            "@id": "xsd:string"
          },
          "sh:path": {
            "@id": "https://good.vibrations.dev/name"
          },
          "sh:minCount": {
            "@value": 1,
            "@type": "xsd:integer"
          },
          "sh:message": "A PromptTemplate must have exactly one name."
        }
      ]

    }

    ]
}
"""


shapes = rdflib.Graph()
shapes.parse(data=shapesJson, format="json-ld")
```

## Data

```python
dataJson = """
{
"@context": {
    "sh": "http://www.w3.org/ns/shacl#",
    "xsd": "http://www.w3.org/2001/XMLSchema#"
  },
  "@graph": [
  {
      "@id": "https://good.vibrations.dev/#prompt-next-question",
      "@type": "https://good.vibrations.dev/PromptTemplate",
      "https://good.vibrations.dev/name": []
    }
  ]
  }


"""

data = rdflib.Graph()
data.parse(data=dataJson, format="json-ld")
```

## Validation

```python
conforms, results_graph, results_text = validate(
    data, shacl_graph=shapes
)
```
